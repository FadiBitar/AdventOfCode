```python
# Day 8 Part 1

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        raw_data = [d.split('|')[1] for d in raw_data]
        data = [d.split() for d in raw_data]
        return data

def count_unique_combinations(data):
    # 0: 6
    # 1: 2 *
    # 2: 5
    # 3: 5
    # 4: 4 *
    # 5: 5
    # 6: 6
    # 7: 3 *
    # 8: 7 *
    # 9: 6
    total = 0
    for row in data:
        lengths = [len(i) for i in row]
        total = total + lengths.count(2) + lengths.count(3) + lengths.count(4) + lengths.count(7)
    return total

# Test first
test_data = read_input_data('test_data/day8.txt')
print("Running on test data. Verifying result...")
print(count_unique_combinations(test_data) == 26)
print('')

# Now run on actual data
real_data = read_input_data('data/day8.txt')
result = count_unique_combinations(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 8 Part 2

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        in_data = [d.split('|')[0] for d in raw_data]
        out_data = [d.split('|')[1] for d in raw_data]
        in_data = [d.split() for d in in_data]
        out_data = [d.split() for d in out_data]
        return in_data, out_data

def contains_all_characters(test_str, ref_str):
    val = True
    for i in range(0, len(ref_str)):
          val = val and ref_str[i] in test_str
    return val
    
def get_decode_key(in_data, out_data):
    all_data = []
    all_data.extend(in_data)
    all_data.extend(out_data)
    all_data = [''.join(sorted(i)) for i in all_data]
    
    #first pass
    decode_key = {}
    encode_key = {}
    for i in all_data:
        if len(i) == 2:
            decode_key[i] = 1
            encode_key[1] = i
        elif len(i) == 3:
            decode_key[i] = 7
            encode_key[7] = i
        elif len(i) == 4:
            decode_key[i] = 4
            encode_key[4] = i
        elif len(i) == 7:
            decode_key[i] = 8
            encode_key[8] = i

    # second pass
    for i in all_data:
        if len(i) == 5:
            # could be 2, 3 or 5. Find 3 now, the rest later
            if contains_all_characters(i, encode_key[1]):
                decode_key[i] = 3
                encode_key[3] = i
                
    # third pass
    for i in all_data:
        if len(i) == 6:
            # could be 0, 6, or 9
            if contains_all_characters(i, encode_key[3]):
                decode_key[i] = 9
                encode_key[9] = i
            elif contains_all_characters(i, encode_key[1]):
                decode_key[i] = 0
                encode_key[0] = i
            else:
                decode_key[i] = 6
                encode_key[6] = i
    
    # fourth pass
    for i in all_data:
        if len(i) == 5 and i not in decode_key.keys():
            if contains_all_characters(encode_key[9], i):
                decode_key[i] = 5
                encode_key[5] = i
            else:
                decode_key[i] = 2
                encode_key[2] = i
    
    return decode_key
    
    
def decode_and_get_total(in_data, out_data):
    total = 0
    for i in range(0, len(out_data)):
        decode_key = get_decode_key(in_data[i], out_data[i])
        
        digits = [decode_key[''.join(sorted(k))] for k in out_data[i]]
        strings = [str(dig) for dig in digits]
        a_string = "". join(strings)
        total = total + int(a_string)       
    return total

# Test first
test_in_data, test_out_data = read_input_data('test_data/day8.txt')
print("Running on test data. Verifying result...")
print(decode_and_get_total(test_in_data, test_out_data) == 61229)
print('')

# Now run on actual data
real_in_data, real_out_data = read_input_data('data/day8.txt')
result = decode_and_get_total(real_in_data, real_out_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 9 Part 1
import numpy as np

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        data = [list(row) for row in raw_data]
        return np.array(data, dtype=np.int32)

def get_adjacent_levels(data, row, col):
    levels = []
    if row > 0:
        levels.append(data[row-1][col])
    if col > 0:
        levels.append(data[row][col-1])
    if row < data.shape[0] - 1:
        levels.append(data[row+1][col])
    if col < data.shape[1] - 1:
        levels.append(data[row][col+1])
    return levels
                    

def calculate_risk_level(data):
    risk_level = 0
    for i in range(0, data.shape[0]):
        for j in range(0, data.shape[1]):
            if data[i][j] < min(get_adjacent_levels(data, i, j)):
                risk_level = risk_level + 1 + data[i][j]
    return risk_level

# Test first
test_data = read_input_data('test_data/day9.txt')
print("Running on test data. Verifying result...")
print(calculate_risk_level(test_data) == 15)
print('')

# Now run on actual data
real_data = read_input_data('data/day9.txt')
result = calculate_risk_level(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 9 Part 2

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        data = [list(row) for row in raw_data]
        return np.array(data, dtype=np.int32)

def find_first_unassigned(basins):
    for i in range(0, basins.shape[0]):
        for j in range(0, basins.shape[1]):
            if basins[i][j] == 0:
                return i, j
    return -1, -1

def get_adjacent_indices(data, row, col):
    indices = []
    if row > 0:
        indices.append((row-1, col))
    if col > 0:
        indices.append((row, col-1))
    if row < data.shape[0] - 1:
        indices.append((row+1, col))
    if col < data.shape[1] - 1:
        indices.append((row, col+1))
    return indices

def grow(data, basins, stack, basin_id):
    if len(stack) == 0:
        return

    i, j = stack.pop(0)
    if basins[i][j] == 0:
        basins[i][j] = basin_id
        stack.extend(get_adjacent_indices(data, i, j))
    #print(basins)
    grow(data, basins, stack, basin_id)

def assign_basins(data):
    #print("DATA")
    #print(data)
    n_locations = data.shape[0] * data.shape[1]
    left_to_assign = n_locations
    basins = np.zeros(data.shape)
    basin_sizes = []

    # find all the 9s. totally useless.
    for i in range(0, data.shape[0]):
        for j in range(0, data.shape[1]):
            if data[i][j] == 9:
                basins[i][j] = -1
    left_to_assign = np.count_nonzero(basins == 0)
    
    current_basin_id = 1
    while(left_to_assign > 0):
        i, j = find_first_unassigned(basins)
        stack = [(i,j)]
        #print("Growing")
        grow(data, basins, stack, current_basin_id)
        
        basin_sizes.append(np.count_nonzero(basins == current_basin_id))
        
        current_basin_id = current_basin_id + 1
        left_to_assign = np.count_nonzero(basins == 0)
        
    basin_sizes.sort(reverse=True)
    return basin_sizes[0] * basin_sizes[1] * basin_sizes[2] 
    
# Test first
test_data = read_input_data('test_data/day9.txt')
print("Running on test data. Verifying result...")
print(assign_basins(test_data) == 1134)
print('')

# Now run on actual data
real_data = read_input_data('data/day9.txt')
result = assign_basins(real_data)
print("Puzzle answer is: " + str(result))
```
