```python
# Day 8 Part 2

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        in_data = [d.split('|')[0] for d in raw_data]
        out_data = [d.split('|')[1] for d in raw_data]
        in_data = [d.split() for d in in_data]
        out_data = [d.split() for d in out_data]
        return in_data, out_data

def contains_all_characters(test_str, ref_str):
    val = True
    for i in range(0, len(ref_str)):
          val = val and ref_str[i] in test_str
    return val
    
def get_decode_key(in_data, out_data):
    all_data = []
    all_data.extend(in_data)
    all_data.extend(out_data)
    all_data = [''.join(sorted(i)) for i in all_data]
    
    #first pass
    decode_key = {}
    encode_key = {}
    for i in all_data:
        if len(i) == 2:
            decode_key[i] = 1
            encode_key[1] = i
        elif len(i) == 3:
            decode_key[i] = 7
            encode_key[7] = i
        elif len(i) == 4:
            decode_key[i] = 4
            encode_key[4] = i
        elif len(i) == 7:
            decode_key[i] = 8
            encode_key[8] = i

    # second pass
    for i in all_data:
        if len(i) == 5:
            # could be 2, 3 or 5. Find 3 now, the rest later
            if contains_all_characters(i, encode_key[1]):
                decode_key[i] = 3
                encode_key[3] = i
                
    # third pass
    for i in all_data:
        if len(i) == 6:
            # could be 0, 6, or 9
            if contains_all_characters(i, encode_key[3]):
                decode_key[i] = 9
                encode_key[9] = i
            elif contains_all_characters(i, encode_key[1]):
                decode_key[i] = 0
                encode_key[0] = i
            else:
                decode_key[i] = 6
                encode_key[6] = i
    
    # fourth pass
    for i in all_data:
        if len(i) == 5 and i not in decode_key.keys():
            if contains_all_characters(encode_key[9], i):
                decode_key[i] = 5
                encode_key[5] = i
            else:
                decode_key[i] = 2
                encode_key[2] = i
    
    return decode_key
    
    
def decode_and_get_total(in_data, out_data):
    total = 0
    for i in range(0, len(out_data)):
        decode_key = get_decode_key(in_data[i], out_data[i])
        
        digits = [decode_key[''.join(sorted(k))] for k in out_data[i]]
        strings = [str(dig) for dig in digits]
        a_string = "". join(strings)
        total = total + int(a_string)       
    return total

# Test first
test_in_data, test_out_data = read_input_data('test_data/day8.txt')
print("Running on test data. Verifying result...")
print(decode_and_get_total(test_in_data, test_out_data) == 61229)
print('')

# Now run on actual data
real_in_data, real_out_data = read_input_data('data/day8.txt')
result = decode_and_get_total(real_in_data, real_out_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 9 Part 1
import numpy as np

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        data = [list(row) for row in raw_data]
        return np.array(data, dtype=np.int32)

def get_adjacent_levels(data, row, col):
    levels = []
    if row > 0:
        levels.append(data[row-1][col])
    if col > 0:
        levels.append(data[row][col-1])
    if row < data.shape[0] - 1:
        levels.append(data[row+1][col])
    if col < data.shape[1] - 1:
        levels.append(data[row][col+1])
    return levels
                    

def calculate_risk_level(data):
    risk_level = 0
    for i in range(0, data.shape[0]):
        for j in range(0, data.shape[1]):
            if data[i][j] < min(get_adjacent_levels(data, i, j)):
                risk_level = risk_level + 1 + data[i][j]
    return risk_level

# Test first
test_data = read_input_data('test_data/day9.txt')
print("Running on test data. Verifying result...")
print(calculate_risk_level(test_data) == 15)
print('')

# Now run on actual data
real_data = read_input_data('data/day9.txt')
result = calculate_risk_level(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 9 Part 2

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        data = [list(row) for row in raw_data]
        return np.array(data, dtype=np.int32)

def find_first_unassigned(basins):
    for i in range(0, basins.shape[0]):
        for j in range(0, basins.shape[1]):
            if basins[i][j] == 0:
                return i, j
    return -1, -1

def get_adjacent_indices(data, row, col):
    indices = []
    if row > 0:
        indices.append((row-1, col))
    if col > 0:
        indices.append((row, col-1))
    if row < data.shape[0] - 1:
        indices.append((row+1, col))
    if col < data.shape[1] - 1:
        indices.append((row, col+1))
    return indices

def grow(data, basins, stack, basin_id):
    if len(stack) == 0:
        return

    i, j = stack.pop(0)
    if basins[i][j] == 0:
        basins[i][j] = basin_id
        stack.extend(get_adjacent_indices(data, i, j))
    #print(basins)
    grow(data, basins, stack, basin_id)

def assign_basins(data):
    #print("DATA")
    #print(data)
    n_locations = data.shape[0] * data.shape[1]
    left_to_assign = n_locations
    basins = np.zeros(data.shape)
    basin_sizes = []

    # find all the 9s. totally useless.
    for i in range(0, data.shape[0]):
        for j in range(0, data.shape[1]):
            if data[i][j] == 9:
                basins[i][j] = -1
    left_to_assign = np.count_nonzero(basins == 0)
    
    current_basin_id = 1
    while(left_to_assign > 0):
        i, j = find_first_unassigned(basins)
        stack = [(i,j)]
        #print("Growing")
        grow(data, basins, stack, current_basin_id)
        
        basin_sizes.append(np.count_nonzero(basins == current_basin_id))
        
        current_basin_id = current_basin_id + 1
        left_to_assign = np.count_nonzero(basins == 0)
        
    basin_sizes.sort(reverse=True)
    return basin_sizes[0] * basin_sizes[1] * basin_sizes[2] 
    
# Test first
test_data = read_input_data('test_data/day9.txt')
print("Running on test data. Verifying result...")
print(assign_basins(test_data) == 1134)
print('')

# Now run on actual data
real_data = read_input_data('data/day9.txt')
result = assign_basins(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 10 Part 1

def get_chunk_score(chunk):
    stack = []
    lookup = {'<':'>', '(':')', '[': ']', '{':'}' }
    score = { ')': 3, ']': 57, '}': 1197, '>': 25137 }
    
    for i in range(0, len(chunk)):
        if chunk[i] in "([{<":
            stack.append(chunk[i])
        else:
            # incomplete
            if len(stack) == 0:
                return 0
            else:
                match = stack.pop()
                if lookup[match] != chunk[i]:
                    return score[chunk[i]]
    #legal
    return 0

def get_illegal_input_score(data):
    total = 0
    for chunk in data:
        total = total + get_chunk_score(chunk)
    return total

# Test first
with open('test_data/day10.txt', 'r') as in_file:
    test_data = in_file.read().split('\n')
print("Running on test data. Verifying result...")
print(get_illegal_input_score(test_data) == 26397)
print('')

# Now run on actual data
with open('data/day10.txt', 'r') as in_file:
    real_data = in_file.read().split('\n')
result = get_illegal_input_score(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 10 Part 2

def get_chunk_autocomplete_score(chunk):
    stack = []
    lookup = {'<':'>', '(':')', '[': ']', '{':'}' }
    score = { ')': 1, ']': 2, '}': 3, '>': 4 }
    
    for i in range(0, len(chunk)):
        if chunk[i] in "([{<":
            stack.append(chunk[i])
        else:
            # incomplete
            if len(stack) == 0:
                return 0
            else:
                match = stack.pop()
                if lookup[match] != chunk[i]:
                    #corrupted
                    return 0
    chunk_score = 0
    for i in range(0, len(stack)):
        chunk_score = chunk_score*5 + score[lookup[stack.pop()]]
    return chunk_score
        
def get_autocomplete_winner(data):
    chunk_scores = []
    for chunk in data:
        score = get_chunk_autocomplete_score(chunk)
        if score != 0:
            chunk_scores.append(score)
    chunk_scores.sort()
    return chunk_scores[int(len(chunk_scores)/2)]

    
# Test first
with open('test_data/day10.txt', 'r') as in_file:
    test_data = in_file.read().split('\n')
print("Running on test data. Verifying result...")
print(get_autocomplete_winner(test_data) == 288957)
print('')

# Now run on actual data
with open('data/day10.txt', 'r') as in_file:
    real_data = in_file.read().split('\n')
result = get_autocomplete_winner(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 11 Part 1
import numpy as np

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
        data = [list(row) for row in raw_data]
        return np.array(data, dtype=np.int32)

def get_adjacent_indices(data, row, col):
    indices = []
    rlist = [r for r in [row-1, row, row+1] if r >= 0 and r < data.shape[0]]
    clist = [c for c in [col-1, col, col+1] if c >= 0 and c < data.shape[1]]
    
    for r in rlist:
        for c in clist:
            indices.append((r, c))
    return indices
    
def count_octopus_flashes(data, n_iter):
    n_flashes = 0
    
    for iteration in range(0, n_iter):
        has_flashed = np.ones(data.shape)
        new_state = data + 1
        
        while(np.count_nonzero(new_state * has_flashed > 9) > 0):
            for i in range(0, new_state.shape[0]):
                for j in range(0, new_state.shape[1]):
                    if new_state[i][j] > 9 and has_flashed[i][j]:
                        has_flashed[i][j] = 0
                        for ind in get_adjacent_indices(new_state, i, j):
                            new_state[ind[0]][ind[1]] = new_state[ind[0]][ind[1]] + 1
        
        n_flashes = n_flashes + np.count_nonzero(has_flashed == 0)
        data = new_state * has_flashed
    return n_flashes
    
    
# Test first
test_data = read_input_data('test_data/day11.txt')
print("Running on test data. Verifying result...")
#print(count_octopus_flashes(test_data, 2) == 35)
print(count_octopus_flashes(test_data, 100) == 1656)
print('')

# Now run on actual data
real_data = read_input_data('data/day11.txt')
result = count_octopus_flashes(real_data, 100)
print("Puzzle answer is: " + str(result))
```

```python
# Day 12 Part 1

class graph:
    def __init__(self,edges):
        self.gdict = {}
        for edge in edges:
            self.addEdge(edge)
            
    def getVertices(self):
        return list(self.gdict.keys())

    def __str__(self):
        s = ""
        for k in self.gdict.keys():
            s = s + k + ": " + str(self.gdict[k]) + "\n"
        return s
    
    def addEdge(self, edge):
        self.addVertex(edge[0])
        self.gdict[edge[0]].append(edge[1])
        self.addVertex(edge[1])
        self.gdict[edge[1]].append(edge[0])

    # Add the vertex as a key
    def addVertex(self, vrtx):
        if vrtx not in self.gdict:
            self.gdict[vrtx] = []

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n')
    raw_data = [node.split('-') for node in raw_data]
    return graph(raw_data)

def is_valid_path(data, path_so_far):
    for k in data.getVertices():
        if k.islower() and path_so_far.count(k) > 1:
            return False
    return True
    

def follow_path(data, next_node, path_so_far, valid_paths):
    #print(path_so_far, next_node)
    path_so_far.append(next_node)
    if is_valid_path(data, path_so_far):
        if next_node == 'end':
            valid_paths.append(path_so_far)
            return

        stack = data.gdict[next_node]
        for node in stack:
            follow_path(data, node, path_so_far.copy(), valid_paths) 

def count_all_paths(data):
    valid_paths = []
    follow_path(data, 'start', [], valid_paths)
    return len(valid_paths)

# Test first
print("Running on test data. Verifying result...")
test_data = read_input_data('test_data/day12a.txt')
print(count_all_paths(test_data) == 10)
test_data = read_input_data('test_data/day12b.txt')
print(count_all_paths(test_data) == 19)
test_data = read_input_data('test_data/day12c.txt')
print(count_all_paths(test_data) == 226)
print('')

# Now run on actual data
real_data = read_input_data('data/day12.txt')
result = count_all_paths(real_data)
print("Puzzle answer is: " + str(result))
```

```python
# Day 13 Part 1
import numpy as np

def read_input_data(filename):
    with open(filename, 'r') as in_file:
        raw_data = in_file.read().split('\n\n')
        dots_list = []
        folds = []
        max_x = 0
        max_y = 0
        
        for d in raw_data[0].split('\n'):
            dot = d.split(',')
            new_dot = (int(dot[0]), int(dot[1]))
            dots_list.append(new_dot)
            max_x = max(max_x, new_dot[0])
            max_y = max(max_y, new_dot[1])
            
        dots = np.zeros((max_x + 1, max_y + 1))
        for d in dots_list:
            dots[d[0]][d[1]] = 1
            
        for f in raw_data[1].split('\n'):
            axis = 0 if f[11] == 'x' else 1
            folds.append((axis, int(f[13:])))
    return dots, folds

def count_dots(dots, folds):
    for fold in folds:
        axis = fold[0]
        index = fold[1]
        
        if axis == 0:
            up_dots = dots[:index, :]
            down_dots = dots[index+1:, :]
            dots = np.logical_or(up_dots, np.flip(down_dots, axis)).astype(int)
            
        if axis == 1:
            left_dots = dots[:,:index]
            right_dots = dots[:,index+1:]
            dots = np.logical_or(left_dots, np.flip(right_dots, axis)).astype(int)
    return np.count_nonzero(dots == 1)

# Test first
test_dots, test_folds = read_input_data('test_data/day13.txt')
print("Running on test data. Verifying result...")
print(count_dots(test_dots, test_folds[:1]) == 17)
print(count_dots(test_dots, test_folds[:2]) == 16)
print('')

real_dots, real_folds = read_input_data('data/day13.txt')
result = count_dots(real_dots, real_folds[:1])
print("Puzzle answer is: " + str(result))
```
